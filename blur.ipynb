```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
image = cv2.imread('giraffe.jpg')

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image_grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
kernel_s = 10
avg_blur = cv2.blur(image_grey,(kernel_s,kernel_s),0)
plt.imshow(avg_blur,cmap='gray')


```

```python

kernel_s =  15
gaussian_blur = cv2.GaussianBlur(image_grey,(kernel_s,kernel_s),0)


plt.imshow(gaussian_blur,cmap='gray')
```

```python

kernel_s =  15
med_blur = cv2.medianBlur(image_grey,3)


plt.imshow(med_blur,cmap='gray')
```
