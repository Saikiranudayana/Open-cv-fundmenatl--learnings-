```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
height,width = 500,600
blue = (255,127,0)
canvas = np.full((height,width,3),blue,dtype=np.uint8)


plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

```python
##Draw a line 

cv2.line(canvas,(50,50),(650,450),(255,255,255),5)

plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

```python
cv2.rectangle(canvas,(100,100),(300,300),(0,255,255),5)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

```python
cv2.circle(canvas,(500,150),100,(0,255,255),5)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

```python
cv2.ellipse(canvas,(350,350),(100,50),45,0,360,(0,255,255),5)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```
