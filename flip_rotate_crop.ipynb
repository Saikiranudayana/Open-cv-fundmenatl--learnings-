```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('giraffe.jpg')
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
flip_horizontal  = cv2.flip(image_rgb ,1)
flip_veritcal  = cv2.flip(image_rgb ,0)
flip_both   = cv2.flip(image_rgb ,-1)


titles = ['Original Image', 'Flip Horizontal', 'Flip Vertical', 'Flip Both']

images = [image_rgb, cv2.cvtColor(flip_horizontal,cv2.COLOR_BGR2RGB), 
          cv2.cvtColor(flip_veritcal,cv2.COLOR_BGR2RGB), 
          cv2.cvtColor(flip_both,cv2.COLOR_BGR2RGB), ]

plt.figure(figsize=(12, 8))

for i in range(4):
    plt.subplot(2, 2, i + 1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
rot_90 = cv2.rotate(image_rgb, cv2.ROTATE_90_CLOCKWISE)
rot_180 = cv2.rotate(image_rgb, cv2.ROTATE_180)
rot_90_anti_clockwise = cv2.rotate(image_rgb, cv2.ROTATE_90_COUNTERCLOCKWISE) 


titles = ['Original Image', 'Rotate 90 Clockwise', 
          'Rotate 180', 'Rotate 90 Anti-Clockwise']

images = [image_rgb,rot_90, rot_180, 
          rot_90_anti_clockwise 
          ]


plt.figure(figsize=(12, 8))

for i in range(4):
    plt.subplot(2, 2, i + 1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
