```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
image = cv2.imread('giraffe.jpg')
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
hist ,bins = np.histogram(image_grey.flatten(), bins=256, range=(0, 256) )


# equalizing the histogram 
equalized_image = cv2.equalizeHist(image_grey)

equalkized_hist ,bins = np.histogram(equalized_image.flatten(), bins=256, range=(0, 256) )


plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(image_grey,cmap='gray')
plt.title('original image') 


plt.subplot(2, 2, 2)
plt.plot(hist, color='blue')
plt.title('original histogram') 


plt.subplot(2, 2, 3)
plt.imshow(equalized_image, cmap='gray')
plt.title('equalized image') 


plt.subplot(2, 2, 4)
plt.plot(equalkized_hist, color='green')
plt.title('equalized histogram') 

plt.tight_layout()
plt.show()
```

```python
image_yuv = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)

image_yuv[:, :, 0] = cv2.equalizeHist(image_yuv[:, :, 0])

equalised_color_image = cv2.cvtColor(image_yuv, cv2.COLOR_YUV2BGR)

equalised_color_image_RGB = cv2.cvtColor(equalised_color_image, cv2.COLOR_BGR2RGB)

# plot the results
plt.figure(figsize=(12, 8))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original image')

plt.subplot(1, 2, 2)
plt.imshow(equalised_color_image_RGB)
plt.title('Equalised rgb image')

plt.tight_layout()
plt.show()
```
