```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
image = cv2.imread('filter.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(image,cmap='gray')
```

```python
horizontal_kernel = np.array([[-1,-1,-1],
                              [0,0,0],
                              [1,1,1]])

horizontal_filterd_image = cv2.filter2D(image,-1,horizontal_kernel)

plt.imshow(horizontal_filterd_image,cmap='gray')
```
